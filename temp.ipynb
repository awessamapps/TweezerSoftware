```python
import matplotlib.pyplot as plt
import numpy as np

# Generate some random values for the scatter plot
p2i = np.random.rand(10)  # Initial values for p2
p1f = np.random.rand(10)  # Final values for p1
p0f = np.random.rand(10)  # Final values for p0


```

```python
plt.figure()
plt.rc('text', usetex=True)  # Enable LaTeX
plt.rc('font', family='serif')  # Use serif font

# Scatter plots with LaTeX labels
plt.scatter(p2i, p1f, label="$p_1$" + "final")
plt.scatter(p2i, p0f, label='$p_0$' + 'final')

# Add vertical line with LaTeX label
plt.axvline(x=0.05, color='r', linestyle='--', label="Phase transition")

# Add legend and LaTeX axis labels
plt.legend()
plt.xlabel("$p_2$" + ",initial")
plt.ylabel("$p_{\\text{final}}$")

# Show the plot
plt.show()
```
